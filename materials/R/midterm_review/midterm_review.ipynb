# classification 1

```r
mtcars
```

```r
mycars <- mtcars |>
    mutate(am_f = factor(am)) # why do we need to do this
mycars
```

```r
# fit a raw model on everythin
# use knn

library(tidymodels)
```

```r
knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = 5) |>
    set_engine("kknn") |> # libarary(kknn)
    set_mode("classification")
knn_spec
```

```r
car_recipe <- recipe(am_f ~ mpg + hp, data = mycars) |>
    step_scale(all_predictors()) |> # why do you need these steps?
    step_center(all_numeric_predictors())
car_recipe
```

```r
knn_fit <- workflow() |>
    add_recipe(car_recipe) |>
    add_model(knn_spec) |>
    fit(data = mycars)

knn_fit
```

```r
cars_with_prediction <- bind_cols(mycars, predict(knn_fit, mycars))
cars_with_prediction
```

```r
cars_with_prediction |>
    count(am_f, .pred_class)
```

```r
cars_with_prediction |>
    group_by(am_f, .pred_class) |>
    summarize(n = n())
```
